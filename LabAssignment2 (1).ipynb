```python
Is_delivering_now_rating = final_df[final_df['Is delivering now']=='Yes'].groupby('Restaurant Name').size().reset_index().rename(columns={0:'Aggregate rating'})
Is_delivering_now_rating
```

```python
# correlation analysis
numerical_columns = ['Average Cost for two', 'Aggregate rating', 'Votes']

# Calculate correlation coefficients
correlation_matrix =final_df[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis',fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Rating Analysis:
df1 =final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
df1

fig = px.bar(df1,
                   x='Rating color',
                   #marginal='box',
                   color='Rating text',
             y='Rating Count',
                   color_discrete_sequence=['Grey', 'Red','Orange','Yellow','Green',"Darkgreen"],
                   title='')
fig.update_layout(bargap=0.1)
fig.show()

votes_rating =  final_df.groupby('Votes')['Aggregate rating'].size().reset_index().rename(columns={0:'Rating Count'})
votes_rating
# 'Aggregate rating' and 'Votes' columns
aggregate_rating = final_df['Aggregate rating']
votes = final_df['Votes']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(votes, aggregate_rating, alpha=0.5, color='yellow')
plt.title('Relationship between Aggregate Ratings and Votes')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()
```

```python
#Country Analysis

from plotly.offline import init_notebook_mode, plot, iplot
import plotly.graph_objs as go
plt.figure(figsize=(12,6))
# import plotly.plotly as py

labels = list(final_df.Country.value_counts().head(20).index)
values = list(final_df.Country.value_counts().head(20).values)

fig = {
    "data":[
        {
            "labels" : labels,
            "values" : values,
            "hoverinfo" : 'label+percent',
            "domain": {"x": [0, .9]},
                        "hole" : 0.6,
            "type" : "pie",
            "rotation":120,
        },
    ],
    "layout": {
        "title" : "Zomato's Presence Countrywise",
        "annotations": [
            {
                "font": {"size":20},
                "showarrow": True,
                "text": "Countries",
                "x":0.2,
                "y":0.9,
            },
        ]
    }
}
iplot(fig);
```
